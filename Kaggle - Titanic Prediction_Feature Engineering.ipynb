```python
#convert cat_vars to ordinal labels
def replace_ordinal_labels(df, var, target='Survived'):
    
    ordered_labels = df.groupby([var])[target].mean().sort_values().index
    
    ordinal_labels = {k: i for i,k in enumerate(ordered_labels,0)}
    
    df[var] = df[var].map(ordinal_labels)
    
```
